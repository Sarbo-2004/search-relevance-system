```python
import pandas as pd
import json
```

```python
data = []
with open("../data/meta_Electronics.jsonl", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if i > 50000:  # Load only first 10,000 rows
            break
        try:
            record = json.loads(line)
            data.append(record)
        except json.JSONDecodeError:
            continue
```

```python
df = pd.DataFrame(data)
```

```python
df = df.dropna(subset=["main_category","price","store"])
```

```python
df = df.drop(columns=["bought_together","subtitle","author"], axis=True)
```

```python
df = df.drop(columns=["images","videos"],axis=True)
```

```python
df = df.reset_index()
```

```python
for col in df.select_dtypes(include='object').columns:
    
    df[col] = df[col].astype(str)
```

```python
df_nlp = df.copy()
```

```python
df_nlp.head()
```

```python
df=df.drop(columns=['index'],axis=True)
df_nlp=df_nlp.drop(columns=['index'],axis=True)
```

```python
import ast
def safe_eval(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return val

# Convert relevant columns
columns_to_parse = ["features", "description", "categories", "details"]
for col in columns_to_parse:
    if col in df_nlp.columns:
        df_nlp[col] = df_nlp[col].apply(safe_eval)
```

```python
def list_to_str(val):
    if isinstance(val, list):
        return ' | '.join(str(x) for x in val)
    return str(val)

def dict_to_str(val):
    if isinstance(val, dict):
        return ' | '.join(f"{k}: {v}" for k, v in val.items())
    return str(val)
```

```python
df_nlp['features'] = df_nlp['features'].apply(list_to_str)
df_nlp['description'] = df_nlp['description'].apply(list_to_str)
df_nlp['categories'] = df_nlp['categories'].apply(list_to_str)
```

```python
df_nlp['details'] = df_nlp['details'].apply(dict_to_str)
```

```python
for col in df_nlp.select_dtypes(include='object').columns:
    df_nlp[col] = df_nlp[col].astype(str)
```
